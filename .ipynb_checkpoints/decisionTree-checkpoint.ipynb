```python
# import packages
import numpy as np
import pandas as pd
```

```python
# load data and examine the data
df = pd.read_csv('notself_employed.csv')
# drop index column
df = df.drop(['Unnamed: 0'], axis=1)
shape = df.shape
print(shape[0], ' observations\n', shape[1], ' attributes')
df.head(10)
df = df[df['Do you currently have a mental health disorder?']!=1]
shape=df.shape
print(shape[0], ' observations\n', shape[1], ' attributes')
df
```

```python
# split the two possible target attribute from the feature attribute
y_current = df['Do you currently have a mental health disorder?']
y_past = df['Have you had a mental health disorder in the past?']
xFeat = df.drop(['Do you currently have a mental health disorder?'], axis = 1)#include the top 2 features

#donot include the top 2 features
#xFeat = df.drop(['Do you currently have a mental health disorder?','Have you had a mental health disorder in the past?','Have you been diagnosed with a mental health condition by a medical professional?'], axis = 1)
print(y_current.head(10))
print(y_past.head(10))
shape = xFeat.shape
print(shape[0], ' observations\n', shape[1], ' attributes')
xFeat.head(10)
```

```python
#decision tree
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
scaler = StandardScaler()
scaler.fit(xFeat)
xFeat_normalized = scaler.transform(xFeat)
xFeat_normalized
```

```python
from sklearn import preprocessing
from sklearn import utils
lab_enc = preprocessing.LabelEncoder()
encoded = lab_enc.fit_transform(y_current)
encoded
```

```python
from matplotlib import pyplot
#initialize decision tree classifier,also tune here
model=DecisionTreeClassifier(min_samples_leaf=25,criterion='gini')
model.fit(xFeat_normalized,encoded)
importance = model.feature_importances_
print(importance)
columnName={}
for i,v in enumerate(importance):
    columnName[xFeat.iloc[:,i].name]=v
# plot feature importance
columnName=dict(sorted(columnName.items(),key=lambda item:item[1],reverse=True))
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.xlabel("Feature Number")
pyplot.ylabel("Importance")
pyplot.show()
print(columnName)
for c in columnName:
    print(c)
    print(columnName[c])
    print("-----")
```

```python

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_validate
cross_val_score(model,xFeat_normalized,encoded,cv=5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_validate
from sklearn import tree
from pprint import pprint
from sklearn import svm
#evaluation
X_train,X_test,y_train,y_test = train_test_split(xFeat_normalized,encoded,test_size=0.3,random_state=0)

model.fit(X_train,y_train)
pre=model.predict(X_test)
labels=[0,1]
print(classification_report(y_test, pre, labels=labels))

# depth = []
# clf = svm.SVC(kernel='linear', C=1).fit(X_train, y_train)
# clf.score(X_test, y_test)
# clf = svm.SVC(kernel='linear', C=1, random_state=0)
# scoring = ['precision_macro', 'recall_macro']
# scores = cross_validate(clf, xFeat_normalized, encoded, scoring=scoring)
# scores
```
