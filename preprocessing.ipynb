```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("mental_health.csv")
data
```

```python
companydata=data.loc[data['Are you self-employed?']==0]
companySize={"1-5":1,"6-25":2,"26-100":3,"100-500":4,"500-1000":5,"More than 1000":6}
result=[]
for row in companydata['How many employees does your company or organization have?']:
    if row==None:
        result.append(0)
    else:
        result.append(companySize[row])
companydata['How many employees does your company or organization have?']=result
```

```python
dropColumns=['Are you self-employed?','Is your primary role within your company related to tech/IT?','Do you have medical coverage (private insurance or state-provided) which includes treatment of  mental health issues?','Do you know local or online resources to seek help for a mental health disorder?','If you have a mental health issue, do you feel that it interferes with your work when being treated effectively?',"If you have a mental health issue, do you feel that it interferes with your work when NOT being treated effectively?",
            'Was your anonymity protected if you chose to take advantage of mental health or substance abuse treatment resources with previous employers?','If you have been diagnosed or treated for a mental health disorder, do you ever reveal this to clients or business contacts?',
            'Do you think that discussing a mental health disorder with previous employers would have negative consequences?','Do you think that discussing a physical health issue with previous employers would have negative consequences?','Would you have been willing to discuss a mental health issue with your previous co-workers?','Would you have been willing to discuss a mental health issue with your direct supervisor(s)?','If you have revealed a mental health issue to a client or business contact, do you believe this has impacted you negatively?',
             'If you have been diagnosed or treated for a mental health disorder, do you ever reveal this to coworkers or employees?','If you have revealed a mental health issue to a coworker or employee, do you believe this has impacted you negatively?',
            'Did you feel that your previous employers took mental health as seriously as physical health?','Do you believe your productivity is ever affected by a mental health issue?','If yes, what percentage of your work time (time performing primary or secondary job functions) is affected by a mental health issue?','Have your previous employers provided mental health benefits?',
            'Were you aware of the options for mental health care provided by your previous employers?','Did your previous employers ever formally discuss mental health (as part of a wellness campaign or other official communication)?',
            'Did your previous employers provide resources to learn more about mental health issues and how to seek help?','Was your anonymity protected if you chose to take advantage of mental health or substance abuse treatment resources with previous employers?','Why or why not?','How willing would you be to share with friends and family that you have a mental illness?',
            'Have your observations of how another individual who discussed a mental health disorder made you less likely to reveal a mental health issue yourself in your current workplace?','If yes, what condition(s) have you been diagnosed with?','If maybe, what condition(s) do you believe you have?','If so, what condition(s) were you diagnosed with?','Have you ever sought treatment for a mental health issue from a mental health professional?',
            'What US state or territory do you live in?','What US state or territory do you work in?','Which of the following best describes your work position?',"Why or why not?.1",
            'Do you know the options for mental health care available under your employer-provided coverage?','Did you hear of or observe negative consequences for co-workers with mental health issues in your previous workplaces?','What country do you live in?']
```

```python
companydata=companydata.drop(columns=dropColumns)
```

```python
# df.groupby('param')['column'].nunique().sort_values(ascending=False).unique().tolist()
YesNoEncode={'Yes':1,'No':0,0:0,1:1}
maybeEncode={'Yes':2,'No':0,'Maybe':1,"I don't know":1,'Never':0,'Sometimes':1,'Always':2,2:2,0:0,1:1}
for col in companydata:
#     print(companydata[col].unique())
    result=[]
    
    if len(companydata[col].unique())==3:
        for row in companydata[col]:
            if row==None:
                result.append(0)
            else:
                result.append(maybeEncode[row])
        companydata[col]=result
    elif len(companydata[col].unique())==2:
        for row in companydata[col]:
            if row==None:
                result.append(0)
            else:
                result.append(YesNoEncode[row])
        companydata[col]=result
companydata
```

```python
#Does your employer provide mental health benefits as part of healthcare coverage?
result=[]
encodeCoverage={'Yes':2,'No':0,"I don't know":1,"Not eligible for coverage / N/A":0,"Not eligible for coverage / N/A":0}
for row in companydata["Does your employer provide mental health benefits as part of healthcare coverage?"]:
    if row==None:
        result.append(0)
    else:
        result.append(encodeCoverage[row])
companydata['Does your employer provide mental health benefits as part of healthcare coverage?']=result
```

```python
#If a mental health issue prompted you to request a medical leave from work, asking for that leave would be:
difficulties={"Very easy":1,"Somewhat easy":2,"Neither easy nor difficult":3,"I don't know":3,"Somewhat difficult":4,"Very difficult":5}
result=[]
for row in companydata['If a mental health issue prompted you to request a medical leave from work, asking for that leave would be:']:
    if row==None:
        result.append(0)
    else:
        result.append(difficulties[row])
companydata['If a mental health issue prompted you to request a medical leave from work, asking for that leave would be:']=result
```

```python
import re
yes=re.compile(r'Yes')
no=re.compile(r'[Nn]o')
maybe=re.compile(r'[Mm]aybe')
```

```python
#Do you feel that being identified as a person with a mental health issue would hurt your career?
uncleanMaybe=['Do you feel that being identified as a person with a mental health issue would hurt your career?','Do you think that team members/co-workers would view you more negatively if they knew you suffered from a mental health issue?','Have you observed or experienced an unsupportive or badly handled response to a mental health issue in your current or previous workplace?']
hurt={"Maybe":1,"Yes":2,"No":0}

for col in uncleanMaybe:
    result=[]
    print(col)
    for row in companydata[col]:
        if pd.isna(row):
            result.append(0)
        else:
            y=yes.match(row)
            n=no.match(row)
            m=maybe.match(row)
            if m:
                result.append(1)
            elif y:
                result.append(2)
            else:
                result.append(0)
    companydata[col]=result
companydata
```

```python
#What country do you work in?
result=[]
for row in companydata['What country do you work in?']:
    if row=='United States of America':
        result.append(1)
    else:
        result.append(0)
companydata['What country do you work in?(1=United States of America, 0=Other)']=result
```
